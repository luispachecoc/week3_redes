```python
import networkx as nx

# create graph
G = nx.Graph()

# create nodes and edges
G.add_edges_from([
    (1, 2), (1, 5), (1, 7),  # Nodo 1 influye en varios actores
    (2, 3), (2, 5), (2, 8),  # Nodo 2 actúa como intermediario
    (3, 4), (3, 9),  # Nodo 3 influye en nodos intermedios y periféricos
    (4, 5), (4, 6), (4, 10),  # Nodo 4 conecta a actores periféricos y centrales
    (5, 6), (5, 8), (5, 10),  # Nodo 5 es un nodo altamente conectado
    (6, 11), (6, 12),  # Nodo 6 influye en actores periféricos
    (7, 2), (7, 8),  # Nodo 7 tiene una influencia en los actores intermediarios
    (8, 9), (8, 11),  # Nodo 8 conecta con los nodos de influencia media y baja
    (9, 10), (9, 12),  # Nodo 9 conecta con actores finales
    (10, 11), (10, 12),  # Nodo 10 actúa como puente entre varios actores
    (11, 12) ]) # Nodo 11 influye en el nodo 12
```

```python
# dibujando
nx.draw(G=G,
        with_labels=True,
        node_color='yellow',
        edgecolors='black')
```
